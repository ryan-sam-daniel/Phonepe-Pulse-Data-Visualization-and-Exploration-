```python
import pandas as pd
import json
import os
import mysql.connector

# Define the path to the data
path = "content/pulse/data/aggregated/transaction/country/india/state/"

# Get the list of states in India
state_list = os.listdir(path)

# Initialize a dictionary to store the data
data = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

# Iterate over each state directory
for state in state_list:
    state_path = os.path.join(path, state)
    if not os.path.isdir(state_path):
        continue
    # Iterate over each year directory within the state
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        if not os.path.isdir(year_path):
            continue
        # Iterate over each quarter JSON file within the year
        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)
            if not quarter_file.endswith('.json'):
                continue
            try:
                # Open and load the JSON data
                with __builtins__.open(quarter_path, 'r') as file:
                    quarter_data = json.load(file)
                
                # Extract and append data to the dictionary
                for transaction in quarter_data['data']['transactionData']:
                    data['State'].append(state)
                    data['Year'].append(year)
                    data['Quarter'].append(int(quarter_file.strip('.json')))
                    data['Transaction_type'].append(transaction['name'])
                    data['Transaction_count'].append(transaction['paymentInstruments'][0]['count'])
                    data['Transaction_amount'].append(transaction['paymentInstruments'][0]['amount'])
            except (FileNotFoundError, json.JSONDecodeError) as e:
                print(f"Error processing file {quarter_path}: {e}")
                continue

# Create a DataFrame from the collected data
Agg_Trans = pd.DataFrame(data)

# Display the DataFrame
print(Agg_Trans.head())
```
